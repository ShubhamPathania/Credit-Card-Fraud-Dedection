```python
from sklearn.preprocessing import MinMaxScaler
```

```python
df2 = df2.sort_values(by='Time')
```

```python
df2[df2.columns[:-1]]
```

```python
scaler = MinMaxScaler()
scaler.fit(df2[df2.columns[:-1]])
scaled_df = scaler.transform(df2[df2.columns[:-1]])
```

```python
X = scaled_df
```

```python
y= df2[df2.columns[-1]].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
```

```python
model = Sequential()
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x=X_train, 
          y=y_train, 
          epochs=200,
          validation_data=(X_test, y_test), verbose=1, batch_size= 50
          )
```

```python
prediction = model.predict(X_test)
prediction.shape
```

```python
prediction = pd.DataFrame(prediction)
```

```python
prediction.columns = ['Probablity']
```

```python
def probab_cal(x, threshold):
    
    if x>threshold:
        return 1
    elif x<threshold:
        return 0
```

```python
prediction['Result'] = prediction.Probablity.apply(lambda x:probab_cal(x, 0.9))
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_true=y_test, y_pred=prediction.Result)
```

```python
confusion_matrix(y_true=y_test, y_pred=prediction.Result)
```
